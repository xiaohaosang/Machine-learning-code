```python
import numpy as np

def loadDataSet(fileName,delim = '\t'):
    fr =open(fileName)
    stringArr = [line.strip().split(delim) for line in fr.readlines()]
    datArr = [list(map(float,line)) for line in stringArr]
    return np.mat(datArr)
dataMat = loadDataSet("./MLiA_SourceCode/machinelearninginaction/Ch13/testSet.txt")
print(dataMat)
print(dataMat.shape)
```

```python
def pca(dataMat, topNfeat=9999999):
    meanVals = np.mean(dataMat, axis=0)   
    # meanVals = np.mat(np.zeros([dataMat.shape[0],dataMat.shape[1]]))  
    meanRemoved = dataMat - meanVals  #其实不需要减meanVals，直接用原始数据
    covMat = np.cov(meanRemoved, rowvar=0)
    print("covMat:\n",covMat)
    eigVals,eigVects = np.linalg.eig(np.mat(covMat))
    print("eigVals:\n",eigVals,"\n","eigVects:\n",eigVects)
    eigValInd = np.argsort(eigVals)
    print("eigValInd:\n",eigValInd)
    eigValInd = eigValInd[:-(topNfeat+1):-1]
    print("eigValInd:\n",eigValInd)
    redEigVects = eigVects[:,eigValInd]
    print("redEigVects:\n",redEigVects)
    print(redEigVects.T*redEigVects)  #u(i).T*u(i)=1
    lowDDataMat = meanRemoved*redEigVects
    reconMat = (lowDDataMat * redEigVects.T) + meanVals 
    return lowDDataMat,reconMat
```

```python
lowMat,reconMat = pca(dataMat,1)
print("lowMat:",lowMat)
print("reconMat:",reconMat)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
test_data = dataMat[:,0].flatten()
print(test_data.shape)
print(type(test_data))
print(type(test_data.A)) # python中一个matrix矩阵名.A 代表将 矩阵转化为array数组类型
print(test_data.A.shape)  
ax.scatter(dataMat[:,0].flatten().A[0],dataMat[:,1].flatten().A[0],marker='^',s=90)
ax.scatter(reconMat[:,0].flatten().A[0],reconMat[:,1].flatten().A[0],marker='o',s=50,c='red')
plt.show()
```
